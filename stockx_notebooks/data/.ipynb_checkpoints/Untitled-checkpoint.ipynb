```python
import pandas as pd
import matplotlib.pyplot as plt

import json
```

```python
with open("./cloud_05-20-2017/adidas-yeezy-boost-350-v2-cream-white.json", "rb") as f :
    data = json.load(f)
```

```python
all_sales_history = []
for bar in data['history'] :
    sales_bar = bar['sales']
    all_sales_history.extend(list(reversed(sales_bar)))
```
